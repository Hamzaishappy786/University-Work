```python
import random
class Account:
    #inititializing attributes in a constructor
    def __init__(self, account_holder, initial_balance):
        self.account_number = random.randint(1000, 9999)
        self.account_holder = account_holder
        self.balance = initial_balance
        self.transactions = []
    #for depositing money into the account
    def deposit(self, amount):
        self.balance = self.balance + amount
        self.transactions.append(f"Deposit: +Rs{amount}.")
    #for withdrawing money from account
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds.")
        else:
            self.balance -= amount
            self.transactions.append(f"Withdrawal: -Rs{amount}")
    #getter for balance
    def get_balance(self):
        return self.balance
    #getter for transactions
    def get_transactions(self):
        return self.transactions
    #string representation of accounts
    def __str__(self):
        return f"Account Holder: {self.account_holder}\nBalance: {self.balance}\n"

class Bank:
    #constructor initializing dictionary of accounts
    def __init__(self):
        self.accounts = {}
    #for creating accounts
    def create_account(self, account_holder, initial_balance):
        account_number = len(self.accounts) + 1
        new_account = Account(account_holder, initial_balance)
        self.accounts[account_number] = new_account
        return account_number
    #for displaying accounts in records
    def display_accounts(self):
        print("Accounts:")
        for acc_num, acc in self.accounts.items():
            print(f"{acc_num}: {acc.account_holder}")
    #getter for account
    def get_account(self, account_number):
        return self.accounts.get(account_number, None)
    #string representation of the bank
    def __str__(self):
        result = ""
        for account_number, account in self.accounts.items():
            result += f"Account {account_number}:\n{str(account)}\n"
        return result

class Transaction:
    #constructor of account number and amount
    def __init__(self, account_number, amount):
        self.account_number = account_number
        self.amount = amount
    #excecuting transactions for specific accounts
    def execute_transaction(self, bank):
        account = bank.get_account(self.account_number)
        if account is not None:
            account.withdraw(self.amount)
            account.transactions.append(f"Transaction: -Rs{self.amount}")
            print("Transaction successful.")
        else:
            print("Invalid account number.")

bank = Bank()   #object bank created
#menu-driven structure
while True:
    print("\n1. Create account")
    print("2. Perform Transaction")
    print("3. Deposit Money")
    print("4. Display accounts")
    print("5. View account information")
    print("0. Exit\n")
    
    choice = int(input("Enter your choice: "))
    if choice == 1:   #creating account
        account_holder = str(input("Enter account holder's name: "))
        initial_balance = int(input("Enter initial balance: "))
        account_number = bank.create_account(account_holder, initial_balance)
        print(f"Account created successfully.\nAccount number: {account_number}")

    elif choice == 2:    #transactions
        account_number = int(input("Enter the account number: "))
        amount = float(input("Enter the transaction amount: "))
        transaction = Transaction(account_number, amount)
        transaction.execute_transaction(bank)
    
    elif choice == 3:    #deposit money in accounts
        bank.display_accounts()
        acc_num = int(input("Enter the account number: "))
        amount = float(input("Enter the deposit amount: "))
        account = bank.get_account(acc_num)
        if account:
            account.deposit(amount)
            print("Deposit money.")
        else:
            print("Invalid information.")
    
    elif choice == 4:   #display all accounts
        bank.display_accounts()

    elif choice == 5:   #for viewing specific information of account
        account_number = int(input("Enter the account number: "))
        account = bank.get_account(account_number)
        if account:
            print(str(account))
            print("Transaction history:")
            for transaction in account.get_transactions():
                print(transaction)
        else:
            print("Invalid account number.")

    elif choice == 0:   #ending the program
        print("Exiting the program. Goodbye!!!")
        break

    else:
        print("Invalid choice. Please enter a number between 0 and 3.")
```

```python
import re
#a) for checking email validity
def email_validity(email):
    pattern_for_valid_email = r"^[a-zA-Z\d._-]+@[a-zA-Z\d.]+\.[a-zA-Z]{2,}$"
    if (bool(re.match(pattern_for_valid_email, email)) == True):
        print(f"\n{email} is a valid mail.")
    else:
        print(f"\n{email} is not a valid mail.")
#b) for checking phone number validity
def ph_number_validity(ph_number):
    pattern_for_phone_number = r"\b(?:\+92\s?)?0?3[0-9]{2}[-\s]?[0-9]{7}\b"
    if (bool(re.match(pattern_for_phone_number, ph_number)) == True):
        print(f"\n{ph_number} is a valid phone number.")
    else:
        print(f"\n{ph_number} is a not valid phone number.")
#c) for checking password validity
def password_validity(password):
    pattern_for_password = r"^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[!#$%^&*_\.]).*$"
    if (bool(re.match(pattern_for_password, password)) == True):
        print(f"\n{password} is a valid password.")
    else:
        print(f"\n{password} is a not valid password.")
        print("Requirements for a strong password:")
        print("At least one uppercase character")
        print("One digit")
        print("One speial character")
#d) url extraction from a sentence
def URL_extractor(URL):
    pattern_for_url = r"[a-z]*\.[A-Za-z\d]*\.[a-z]*"
    print("Text:", URL)
    if bool(re.search(pattern_for_url, URL)):
        print(re.findall(pattern_for_url, URL))
    else:
        print(f"Found nothing.")
        
email = str(input("Enter the email: "))
ph_number = int(input("Enter the phone number: "))
password = str(input("Enter the password: "))
URL = str(input("Enter the text: "))
email_validity(email)
ph_number_validity(ph_number)
password_validity(password)
URL_extractor(URL)
```
