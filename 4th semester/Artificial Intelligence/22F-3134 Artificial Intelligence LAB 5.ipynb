```python
def generateSuccessors(state, capacities):
    successorsList = []

    fillFourLitreJug(state, capacities, successorsList)
    fillThreeLitreJug(state, capacities, successorsList)
    emptyFourLitreJug(state, capacities, successorsList)
    emptyThreeLitreJug(state, capacities, successorsList)
    pour3to4(state, capacities, successorsList)
    pour4to3(state, capacities, successorsList)
    pourAllFrom3to4(state, capacities, successorsList)
    pourAllFrom4to3(state, capacities, successorsList)

    return successorsList

def stateSpaceCreation(capacities):
    stateSpace = {}
    toExplore = [(0, 0)]
    index = 0

    while index < len(toExplore):
        currentState = toExplore[index]
        index += 1

        if currentState in stateSpace:
            continue

        stateSpace[currentState] = generateSuccessors(currentState, capacities)
        toExplore.extend(stateSpace[currentState])

    return stateSpace


def fillFourLitreJug(state, capacities, successorsList):
    if state[0] < capacities[0]:
        successorsList.append((capacities[0], state[1]))

def fillThreeLitreJug(state, capacities, successorsList):
    if state[1] < capacities[1]:
        successorsList.append((state[0], capacities[1]))

def emptyFourLitreJug(state, capacities, successorsList):
    if state[0] > 0:
        successorsList.append((0, state[1]))

def emptyThreeLitreJug(state, capacities, successorsList):
    if state[1] > 0:
        successorsList.append((state[0], 0))

def pour3to4(state, capacities, successorsList):
    if 0 < state[0] + state[1] >= capacities[0] and state[1] > 0:
        successorsList.append((capacities[0], state[1] - (capacities[0] - state[0])))

def pour4to3(state, capacities, successorsList):
    if 0 < state[0] + state[1] >= capacities[1] and state[0] > 0:
        successorsList.append((state[0] - (capacities[1] - state[1]), capacities[1]))

def pourAllFrom3to4(state, capacities, successorsList):
    if 0 < state[0] + state[1] <= capacities[0] and state[1] > 0:
        successorsList.append((state[0] + state[1], 0))

def pourAllFrom4to3(state, capacities, successorsList):
    if 0 < state[0] + state[1] <= capacities[1] and state[0] > 0:
        successorsList.append((0, state[0] + state[1]))

def printStateSpace(stateSpace):
    for key, value in stateSpace.items():
        print(key, ":", value)

def dfsSearch(stateSpace, start, goal, visited=None, path=None):

    if visited is None:
        visited = set()

    if path is None:
        path = []

    visited.add(start)
    path.append(start)

    if start == goal:
        return path

    for neighbor in stateSpace[start]:
        if neighbor not in visited:
            result = dfsSearch(stateSpace, neighbor, goal, visited, path)
            if result:
                return result

    return None

if '__main__' == __name__:

    jugCapacities = (4, 3)

    stateSpace = stateSpaceCreation(jugCapacities)

    start = (0, 0)
    goal = (2, 0)

    printStateSpace(stateSpace)

    visited = set()

    result = dfsSearch(stateSpace, start, goal, visited)

    if result:
        print("DFS Path:", result)
    else:
        print("No path found.")
```
